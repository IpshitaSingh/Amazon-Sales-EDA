# Amazon Sales - Exploratory Data Analysis

This project explores the sales data of products on Amazon, with a focus on understanding the relationship between discounts, ratings, and reviews. The dataset includes information on categories, discounted and actual prices, discount percentages, ratings, rating counts, review titles, and review content. 

The <b>goal</b> of this exploratory data analysis is to gain insights into the factors that impact sales on Amazon and to identify any interesting trends or patterns in the data. By analyzing this dataset, I hope to gain insights that can inform future marketing and sales strategies for businesses selling on Amazon.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/IpshitaSingh/Datasets/main/amazon_sales.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
new_df = df.drop(['product_id', 'product_name', 'about_product', 'user_id', 'user_name', 'review_id', 'img_link', 'product_link'], axis = 1)
```

```python
new_df.columns
```

```python
new_df['discounted_price'] = new_df['discounted_price'].str.replace(',', '').str.replace('₹', '')
new_df['actual_price'] = new_df['actual_price'].str.replace(',', '').str.replace('₹', '')
new_df['discount_percentage'] = new_df['discount_percentage'].str.replace('%', '')
new_df['rating'] = new_df['rating'].str.replace('|', '')
new_df['rating_count'] = new_df['rating_count'].str.replace(',', '')
```

```python
cols = ['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']
new_df[cols] = new_df[cols].apply(pd.to_numeric, errors='ignore')
```

```python
new_df.dtypes
```

```python
new_df.describe()
```

<h4>Insights</h4>

* The average discounted price is ₹3129.28, while the average actual price is ₹5451.07. This suggests that, on average, items are being sold at a discount on Amazon


* The average discount percentage is 47.65%, which is a significant discount


* The average rating is 4.10, with a standard deviation of 0.29, indicating that most items have high ratings


* The minimum discounted price and actual price are the same at ₹39, while the maximum discounted price is ₹77990 and the maximum actual price is ₹139900, suggesting that there is a wide range of prices for items on Amazon


* The median rating is 4.1, which is the same as the mean, indicating a roughly symmetric distribution of ratings


* The median rating count is 5179, which is much lower than the mean, meaning that a few products have a really high number of ratings, but most of the products have a much lower number of ratings.

<h3>Visualization</h3>

My objective is to utilize this dataset for the purpose of obtaining answers to the following questions:

> What are the top-selling product categories on Amazon? 

> What is the average discount percentage on these categories?

> Do products in these categories offer the highest discounts?

> Does the category of a product have an impact on its rating?

Relationship between review title/content and rating: It would be interesting to see if there are any specific keywords or phrases that are commonly used in high-rated reviews.

<h5>What are the top-selling categories on Amazon?</h5>

```python
# Grouping the data by the "category" column and getting the sum of rating count for each category to find popular categories
sales_by_category = new_df.groupby('category')['rating_count'].sum()

# Sorting the series in descending order to get the top 10 categories
top_categories = sales_by_category.sort_values(ascending=False).head(10)

top_categories
```

```python
# Create bar chart
plt.barh(top_categories.index, top_categories.values, color='mediumaquamarine')

plt.style.use('ggplot')

# Set title and labels
plt.title('Top Selling Categories on Amazon', fontsize=18)
plt.xlabel('Total Rating Count (in 100,000)', fontsize=14)
plt.ylabel('Category', fontsize=14)
    
# Display the chart
plt.show()
```

<h5>How much discount is offered on average for each top category?</h5>

```python
avg_discount_by_category = new_df.groupby('category')['discount_percentage'].mean()
```

```python
rating_and_discount = pd.concat([sales_by_category, avg_discount_by_category], axis=1)
rating_and_discount.sort_values('rating_count', ascending=False, inplace=True)
top_categories_discounts = rating_and_discount.head(10)
```

```python
# Plot the top 10 selling categories with their average discount
top_categories_discounts.plot(kind='barh', y='discount_percentage', color='teal', width = 0.8)
 
plt.style.use('ggplot')
    
plt.title('Average Discount by Top Selling Categories', fontsize=18)
plt.xlabel('Average Discount (%)', fontsize = 14)
plt.ylabel('Category', fontsize = 14)

plt.show()
```

<h4>Insights</h4>

* This graph suggests that level of discount doesn't necessarily correlate with the number of sales. The fact that the 1st ranking category doesn't have the highest discount suggests that customers may be more interested in other factors such as the product quality, brand reputation, or unique features.


* The 3rd category's (smart phones and mobile phones) low discount level could indicate that the products in this category are highly desirable and therefore don't require as much discounting to sell well. On the other hand, the 5th category's (smartwatches) high level of discount may suggest that the products are not selling as well and require a discount to entice customers.


* The varying levels of discount among the top 10 categories suggest that the website is using a targeted pricing strategy based on the demand and competition within each category.

<h5>Does the category of a product have an impact on its rating?</h5>

```python
# Create a scatter plot
plt.scatter(new_df['category'], new_df['rating'], c=new_df['rating'], cmap='RdPu',alpha=0.5)

# Set the plot title and axis labels
plt.title("Relationship between Category and Rating")
plt.xlabel("Categories")
plt.ylabel("Rating")

plt.xticks([])

# Show the plot
plt.show()
```

<h4>Insights</h4>

* Products in this dataset seem to be good quality as most ratings are between 4.0 and 4.5


* More products fall in the range of 3.5 to 4 than in the range of 4.5 to 5, suggesting that a larger proportion of products are rated slightly lower than the highest-rated products.


* A limited number of products have either a significantly high or low rating.
